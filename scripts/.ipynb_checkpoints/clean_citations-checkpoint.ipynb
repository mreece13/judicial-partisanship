```python
import pandas as pd
from eyecite import get_citations, annotate_citations
import re
```

```python
d = pd.read_csv('../data/opinions.csv')
d.dropna(subset=['plain_text'], inplace=True)

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

d['clean_text'] = d['plain_text'].apply(lambda text: annotate_citations(text, [[c.span(), "<a>", "</a>"] for c in get_citations(text)]))
d['clean_text'] = d["clean_text"].apply(lambda text: remove_html(remove_urls(text)))
```
